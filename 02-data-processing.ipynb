L do trong Pandas, hm apply hot ng theo c ch Row-wise (tng dng) v n mong i mt gi tr tr v  gn vo ct mi (New_Role)-> khng th xa dng trc tip ngay bn trong hm apply c, v n ang duyt qua dataframe. Nu xa dng khi ang duyt, n s gy li index hoc kt qu sai lch.

```
import pandas as pd
import numpy as np

def strict_role_correction_v3(df):
    
    temp = df.copy()
    
    # 1. TNH IM (SCORING)
    # NaN = 0 im, C d liu (>0) = 1 im
    # (Lu : D liu ca bn  l s hoc NaN)
    
    # Nhm Student (Max 3 im): Academic Pressure, Study Satisfaction, CGPA
    s_cols = ['Academic Pressure', 'Study Satisfaction', 'CGPA']
    temp['S_Score'] = temp[s_cols].notna().sum(axis=1)
    
    # Nhm Worker (Max 2 im): Work Pressure, Job Satisfaction
    w_cols = ['Work Pressure', 'Job Satisfaction']
    temp['W_Score'] = temp[w_cols].notna().sum(axis=1)
    
    # HM LOGIC XT DUYT
    def decide_action(row):
        role = row['Working Professional or Student']
        s = row['S_Score']
        w = row['W_Score']
        
        # LOGIC CHO NHM KHAI L STUDENT 
        if role == 'Student':
            # Case: Gi nguyn (Student)
            # 3-0, 3-1, 2-1, 2-0, 1-0
            if (s == 3) or \
               (s == 2 and w <= 1) or \
               (s == 1 and w == 0):
                return 'Student'
            
            # Case: Ha -> Student (1-1, 2-2)
            if (s == 1 and w == 1) or (s == 2 and w == 2):
                return 'Student'
            
            # Case: i sang Worker (0-2, 1-2)
            if (s == 0 and w == 2) or (s == 1 and w == 2):
                return 'Working Professional'
            
            # Case: Drop (0-1)
            if s == 0 and w == 1:
                return 'DROP'
                
        # LOGIC CHO NHM KHAI L WORKER 
        elif role == 'Working Professional':
            # Case: Gi nguyn (Worker)
            # 2-0, 2-1 (W=2, S=0/1)
            if (w == 2 and s <= 1):
                return 'Working Professional'
                
            # Case: Ha -> Worker (1-1, 2-2)
            if (w == 1 and s == 1) or (w == 2 and s == 2):
                return 'Working Professional'
            
            # Case: i sang Student (0-3, 1-3, 1-2 -> W=0/1, S=3/2)
            if (w == 0 and s == 3) or \
               (w == 1 and s == 3) or \
               (w == 1 and s == 2):
                return 'Student'
            
            # B sung: W=0, S=2 (Tng t 0-3) -> Student
            if w == 0 and s == 2:
                return 'Student'

            # Case: Drop (0-1 -> W=0, S=1)
            if w == 0 and s == 1:
                return 'DROP'
        
        # Cc trng hp cn li (0-0...) -> Gi nguyn
        return role

    # 3. P DNG LOGIC
    temp['New_Role'] = temp.apply(decide_action, axis=1)
    
    # 4. THC HIN THAY I
    
    # Bc A: Bo co
    n_drop = (temp['New_Role'] == 'DROP').sum()
    n_switch = (temp['Working Professional or Student'] != temp['New_Role']) & (temp['New_Role'] != 'DROP')
    print(f"-> S lng dng s xa (DROP): {n_drop}")
    print(f"-> S lng dng s i vai (SWITCH): {n_switch.sum()}")

    # Bc B: Drop dng
    df_clean = df[temp['New_Role'] != 'DROP'].copy()
    temp_clean = temp[temp['New_Role'] != 'DROP'] # ng b temp
    
    # Bc C: Cp nht Role mi
    mask_update = df_clean['Working Professional or Student'] != temp_clean['New_Role']
    df_clean.loc[mask_update, 'Working Professional or Student'] = temp_clean.loc[mask_update, 'New_Role']
    
    # 5. DN DP D LIU THA (CLEANUP)
    # Xa d liu ca phe i lp (v NaN  nht qun vi vic gi missing)
    
    mask_st = df_clean['Working Professional or Student'] == 'Student'
    df_clean.loc[mask_st, ['Work Pressure', 'Job Satisfaction']] = np.nan
    
    mask_wk = df_clean['Working Professional or Student'] == 'Working Professional'
    df_clean.loc[mask_wk, ['Academic Pressure', 'Study Satisfaction', 'CGPA']] = np.nan
    
    return df_clean

data_copy = strict_role_correction_v3(data_copy)
```

```
import pandas as pd
import numpy as np
import re

def process_city_direct(df):
    print("--- BT U X L TRC TIP CT CITY ---")
    
    #  CHUN HA (Xa k t c bit u dng + Title Case)
    def normalize_text(val):
        val = str(val).strip()
        if val.lower() == 'nan': return np.nan
        # Xa mi k t khng phi ch ci  u chui (v d: *Nagpur -> Nagpur)
        val = re.sub(r'^[^a-zA-Z]+', '', val)
        return val.title()

    df['City'] = df['City'].apply(normalize_text)

    # A LI CHNH T (Mapping trc tip)
    typo_map = {
        'Molkata': 'Kolkata', 
        'Tolkata': 'Kolkata',
        'Khaziabad': 'Ghaziabad', 
        'Galesabad': 'Ghaziabad',
        'Less Delhi': 'Delhi',
        'Nalyan': 'Kalyan', 
        'Less Than 5 Kalyan': 'Kalyan',
        'Ishanabad': 'Ghaziabad' 
    }
    df['City'] = df['City'].replace(typo_map)

    # LC D LIU HP L 
    # Ch gi li tn cc thnh ph chun, cn li (Tn ngi, Bng cp...) -> NaN
    valid_cities = [
        'Ahmedabad', 'Agra', 'Bangalore', 'Bhopal', 'Chennai', 'Delhi', 
        'Faridabad', 'Ghaziabad', 'Gurgaon', 'Hyderabad', 'Indore', 
        'Jaipur', 'Kalyan', 'Kanpur', 'Kolkata', 'Lucknow', 'Ludhiana', 
        'Meerut', 'Mumbai', 'Nagpur', 'Nashik', 'Patna', 'Pune', 
        'Rajkot', 'Srinagar', 'Surat', 'Thane', 'Vadodara', 'Varanasi', 
        'Vasai-Virar', 'Visakhapatnam'
    ]
    
    # Bin nhng gi tr khng nm trong list ny thnh NaN
    df.loc[~df['City'].isin(valid_cities), 'City'] = np.nan
    
    # Xa cc dng rc (NaN)
    initial_len = len(df)
    df = df.dropna(subset=['City'])
    print(f"->  loi b {initial_len - len(df)} dng rc trong ct City.")

    return df

# P DNG (Chy trn data_copy)
data_copy = process_city_direct(data_copy)

# Kim tra kt qu
print("\nPhn phi ct City sau khi x l:")
print(data_copy['City'].value_counts())
```

```
def categorize_sleep(val):
    # Chuyn v chui thng  x l
    val = str(val).lower().strip()
    
    # NHM 1: > 8 hours
    # Bao gm: More than 8, 8-9, 9-11, 10-11... 
    if any(x in val for x in ['more than 8', '8-9', '9-11', '10-11']):
        return 'More than 8 hours'
    
    # NHM 2: 6-8 hours (Logic: 6 < hour <= 8) 
    # Bao gm: 7-8, 6-7, 6-8, 8 hours... 
    elif any(x in val for x in ['7-8', '6-7', '6-8', '8 hours']):
        return '6-8 hours'
        
    # NHM 3: <= 6 hours 
    # Bao gm: Less than 5, 5-6, 3-4, 4-5...
    elif any(x in val for x in ['less than 5', '5-6', '3-4', '4-5', '1-6', '4-6', 
                                '2-3', '3-6', 'than 5', '1-2', '1-3']):
        return 'Less than or equal 6 hours'

    # RC (Pune, No, 45...)
    else:
        return np.nan

data_copy['Sleep Duration'] = data_copy['Sleep Duration'].apply(categorize_sleep)

# Drop NaN
data_copy = data_copy.dropna(subset=['Sleep Duration'])

sleep_order = ['Less than or equal 6 hours', '6-8 hours', 'More than 8 hours']
data_copy['Sleep Duration'] = pd.Categorical(data_copy['Sleep Duration'], categories=sleep_order, ordered=True)

print("--- PHN PHI MI ---")
print(data_copy['Sleep Duration'].value_counts().sort_index())
```

```
import pandas as pd
import numpy as np

def group_degree_level(val):
    # Chuyn v ch thng v xa khong trng
    val = str(val).lower().strip()
    
    # NHM TIN S 
    if 'phd' in val:
        return 'Doctorate'
        
    # NHM THC S (Master / Post-Graduate)
    if any(x in val for x in ['m.tech', 'mtech', 'me', 'm.sc', 'msc', 'm.s', 
                              'm.com', 'mcom', 'mba', 'mca', 'm.ed', 'med', 
                              'llm', 'ma', 'm.a', 'm.arch', 'march', 'm.pharm', 'mpharm',
                              'md', 'doctor', 'ca', 'aca','mhm']):
        return 'Master'

    # NHM C NHN (Bachelor / Undergraduate)
    if any(x in val for x in ['b.tech', 'btech', 'b tech', 'be', 'bachelor', 
                              'b.sc', 'bsc', 'b.com', 'bcom', 'bba', 'bb', 'b ba', 
                              'bca', 'mbbs', 'b.pharm', 'bpharm', 'b.ph', 
                              'b.arch', 'barch', 'b.ed', 'bed', 'llb', 'll b', 
                              'ba', 'b.a', 'bhm']):
        return 'Bachelor'

    # NHM TRUNG HC (High School)
    if any(x in val for x in ['class 12', '12th', 'hsc', 'class 11']):
        return 'High School'
        
    # CC TRNG HP RC
    return np.nan

data_copy['Degree'] = data_copy['Degree'].apply(group_degree_level)

# Xa cc dng rc (NaN) trc tip trn ct Degree
# (Nhng gi tr nh 'Plumber', 'Manager'  bin thnh NaN)
data_copy = data_copy.dropna(subset=['Degree'])

print("PHN PHI TRNH  HC VN (4 NHM)")
print(data_copy['Degree'].value_counts())
```

```
import pandas as pd
import numpy as np

# Hm gom nhm ngh nghip (Phin bn gp Unspecified)
def group_profession_domain_final(val):
    val = str(val).lower().strip()
    
    # 1. NHM SINH VIN (Gi nguyn v y l i tng chnh)
    if val == 'student':
        return 'Student'
    
    # --- CC NHM NGNH C TH (Domain Specific) ---
    
    # 2. Education & Research
    if any(x in val for x in ['teacher', 'educational consultant', 'academic', 'researcher']):
        return 'Education/Research'

    # 3. Medical & Health
    if any(x in val for x in ['doctor', 'pharmacist', 'chemist', 'medical']):
        return 'Medical/Health'

    # 4. Tech & IT
    if any(x in val for x in ['data scientist', 'software engineer', 'developer', 'dev']):
        return 'Tech/IT'

    # 5. Business & Finance
    if any(x in val for x in ['manager', 'analyst', 'accountant', 'consultant', 
                              'entrepreneur', 'sales', 'marketing', 'investment banker']):
        return 'Business/Finance'

    # 6. Creative & Media
    if any(x in val for x in ['writer', 'designer', 'marketer', 'digital']):
        return 'Creative/Media'

    # 7. Engineering & Architecture
    if any(x in val for x in ['engineer', 'architect']):
        return 'Engineering/Architecture'

    # 8. Legal
    if any(x in val for x in ['lawyer', 'judge']):
        return 'Legal'

    # 9. Service & Operations
    if any(x in val for x in ['chef', 'pilot', 'support', 'travel', 'electrician', 'plumber']):
        return 'Service/Operations'

    # --- 10. NHM CN LI (GP CHUNG) 
    # Bao gm: 'Unspecified', 'nan', v cc ngh lt vt khng thuc cc nhm trn
    return 'Other Professions'

data_copy['Profession'] = data_copy['Profession'].apply(group_profession_domain_final)

# Kim tra kt qu
print("--- PHN PHI NGH NGHIP MI ( GP UNSPECIFIED) ---")
print(data_copy['Profession'].value_counts())
```

```
from sklearn.impute import KNNImputer
import pandas as pd
import numpy as np

def impute_cgpa_knn_workflow(train_df, test_df):
    print("--- BT U IMPUTE CGPA BNG KNN (THEO WORKFLOW BN YU CU) ---")
    
    # 1. XC NH CT V I TNG
    knn_cols = ['Academic Pressure', 'Study Satisfaction', 'Age', 'CGPA']
    
    role_col = 'Working Professional or Student' 

    # 2. P DNG HARD RULE (Ngi i lm -> 0)
    mask_worker_tr = train_df[role_col] == 1
    mask_worker_te = test_df[role_col] == 1
    
    train_df.loc[mask_worker_tr, 'CGPA'] = 0
    test_df.loc[mask_worker_te, 'CGPA'] = 0
    print("->  gn CGPA = 0 cho Working Professional.")

    # 3. CHY KNN CHO SINH VIN
    # Ch chy trn nhm Student (Role == 0)
    mask_student_tr = train_df[role_col] == 0
    mask_student_te = test_df[role_col] == 0
    
    print("-> ang tm hng xm (KNN)  in im cho Sinh vin...")
    
    imputer = KNNImputer(n_neighbors=5)
    
    # A. Fit trn train Student
    student_train_data = train_df.loc[mask_student_tr, knn_cols]
    imputed_train_data = imputer.fit_transform(student_train_data)
    
    # Gn ngc li vo Train (Ct CGPA l ct cui cng trong list knn_cols)
    cgpa_idx = knn_cols.index('CGPA') 
    train_df.loc[mask_student_tr, 'CGPA'] = imputed_train_data[:, cgpa_idx]
    
    # B. Tranform vo test ca SVien
    # (Dng quy lut  hc t Train  in cho Test)
    if mask_student_te.sum() > 0:
        student_test_data = test_df.loc[mask_student_te, knn_cols]
        imputed_test_data = imputer.transform(student_test_data)
        test_df.loc[mask_student_te, 'CGPA'] = imputed_test_data[:, cgpa_idx]

    # 4. CLIP 
    # m bo khng c im no < 0 hoc > 10
    train_df['CGPA'] = train_df['CGPA'].clip(0, 10)
    test_df['CGPA'] = test_df['CGPA'].clip(0, 10)
    
    #  Lm trn 2 ch s thp phn 
    train_df['CGPA'] = train_df['CGPA'].round(2)
    test_df['CGPA'] = test_df['CGPA'].round(2)

    return train_df, test_df

X_train, X_test = impute_cgpa_knn_workflow(X_train, X_test)

# Kim tra
print(X_train['CGPA'].describe())
```

```
from sklearn.impute import KNNImputer
import pandas as pd
import numpy as np

def impute_likert_knn_numeric_role_v2(train_df, test_df):
    
    # Cu hnh Role
    role_col = 'Working Professional or Student'
    worker_val = 1
    student_val = 0

    imputer = KNNImputer(n_neighbors=5)

    # X L CHO SINH VIN (Student)
    student_features = ['Age', 'CGPA', 'Academic Pressure', 'Study Satisfaction', 'Financial Stress']
    student_targets = ['Academic Pressure', 'Study Satisfaction', 'Financial Stress']
    
    mask_student_tr = train_df[role_col] == student_val
    mask_student_te = test_df[role_col] == student_val
    
    print(f"-> ang x l nhm Student (Role={student_val})...")
    
    if mask_student_tr.sum() > 0:
        data_tr = train_df.loc[mask_student_tr, student_features].copy()
        imputed_tr = imputer.fit_transform(data_tr)
        
        # Gn li (Cn gn Financial Stress nu n b thiu)
        for col in student_targets:
            idx = student_features.index(col)
            train_df.loc[mask_student_tr, col] = imputed_tr[:, idx]
            
        if mask_student_te.sum() > 0:
            data_te = test_df.loc[mask_student_te, student_features].copy()
            imputed_te = imputer.transform(data_te)
            for col in student_targets:
                idx = student_features.index(col)
                test_df.loc[mask_student_te, col] = imputed_te[:, idx]

    # Gn 0 cho Work (Hard Rule)
    train_df.loc[mask_student_tr, ['Work Pressure', 'Job Satisfaction']] = 0
    test_df.loc[mask_student_te, ['Work Pressure', 'Job Satisfaction']] = 0


    # 2. X L CHO NGI I LM (Working Pro)
    worker_features = ['Age', 'Work Pressure', 'Job Satisfaction', 'Financial Stress']
    worker_targets = ['Work Pressure', 'Job Satisfaction', 'Financial Stress']
    
    mask_worker_tr = train_df[role_col] == worker_val
    mask_worker_te = test_df[role_col] == worker_val
    
    print(f"-> ang x l nhm Worker (Role={worker_val})...")
    
    if mask_worker_tr.sum() > 0:
        data_tr = train_df.loc[mask_worker_tr, worker_features].copy()
        imputed_tr = imputer.fit_transform(data_tr)
        
        for col in worker_targets:
            idx = worker_features.index(col)
            train_df.loc[mask_worker_tr, col] = imputed_tr[:, idx]
            
        if mask_worker_te.sum() > 0:
            data_te = test_df.loc[mask_worker_te, worker_features].copy()
            imputed_te = imputer.transform(data_te)
            for col in worker_targets:
                idx = worker_features.index(col)
                test_df.loc[mask_worker_te, col] = imputed_te[:, idx]

    # Gn 0 cho Academic (Hard Rule)
    cols_zero = ['Academic Pressure', 'Study Satisfaction', 'CGPA']
    train_df.loc[mask_worker_tr, cols_zero] = 0
    test_df.loc[mask_worker_te, cols_zero] = 0

    # 3. HU X L (ROUND & CLIP)
    all_likert = ['Academic Pressure', 'Study Satisfaction', 'Work Pressure', 'Job Satisfaction', 'Financial Stress']
    
    for col in all_likert:
        train_df[col] = train_df[col].round().clip(0, 5) # Clip 0-5 an ton, d Financial Stress min l 1
        test_df[col] = test_df[col].round().clip(0, 5)

    print(" KNN Impute cho tt c Likert + Financial Stress!")
    return train_df, test_df

# P DNG
X_train, X_test = impute_likert_knn_numeric_role_v2(X_train, X_test)

# Kim tra
print("\nCheck missing (bao gm Financial Stress):")
print(X_train[['Academic Pressure', 'Work Pressure', 'Financial Stress']].isnull().sum())
```

```
from sklearn.impute import KNNImputer
import pandas as pd
import numpy as np

def impute_age_knn(train_df, test_df):
    print("--- BT U IMPUTE CT AGE BNG KNN ---")
    
    role_col = 'Working Professional or Student'
    worker_val = 1
    student_val = 0
    
    imputer = KNNImputer(n_neighbors=5)
    
    # 1. X L CHO SINH VIN (Student)
    # Dng cc c im ca sinh vin  on tui
    student_features = ['CGPA', 'Academic Pressure', 'Study Satisfaction', 'Financial Stress', 'Age']
    
    mask_student_tr = train_df[role_col] == student_val
    mask_student_te = test_df[role_col] == student_val
    
    print(f"-> ang chy KNN d on Tui cho Student...")
    
    if mask_student_tr.sum() > 0:
        data_tr = train_df.loc[mask_student_tr, student_features].copy()
        
        # Fit & Transform trn Train
        imputed_tr = imputer.fit_transform(data_tr)
        
        # Gn li ct Age
        age_idx = student_features.index('Age')
        train_df.loc[mask_student_tr, 'Age'] = imputed_tr[:, age_idx]
        
        # Transform trn Test
        if mask_student_te.sum() > 0:
            data_te = test_df.loc[mask_student_te, student_features].copy()
            imputed_te = imputer.transform(data_te)
            test_df.loc[mask_student_te, 'Age'] = imputed_te[:, age_idx]

    # 2. X L CHO NGI I LM (Worker)
    # Dng c im ngi i lm  on tui
    worker_features = ['Work Pressure', 'Job Satisfaction', 'Financial Stress', 'Age']
    
    mask_worker_tr = train_df[role_col] == worker_val
    mask_worker_te = test_df[role_col] == worker_val
    
    print(f"-> ang chy KNN d on Tui cho Worker...")
    
    if mask_worker_tr.sum() > 0:
        data_tr = train_df.loc[mask_worker_tr, worker_features].copy()
        imputed_tr = imputer.fit_transform(data_tr)
        
        age_idx = worker_features.index('Age')
        train_df.loc[mask_worker_tr, 'Age'] = imputed_tr[:, age_idx]
        
        if mask_worker_te.sum() > 0:
            data_te = test_df.loc[mask_worker_te, worker_features].copy()
            imputed_te = imputer.transform(data_te)
            test_df.loc[mask_worker_te, 'Age'] = imputed_te[:, age_idx]

    # 3. HU X L (ROUND & CLIP)
    # Tui phi l s nguyn v nm trong khong hp l (10-90)
    for df in [train_df, test_df]:
        df['Age'] = df['Age'].round().clip(10, 90).astype(int)

    print(" KNN Impute cho ct Age!")
    return train_df, test_df

# P DNG
X_train, X_test = impute_age_knn(X_train, X_test)

# Kim tra kt qu
print("\nKim tra missing Age:")
print(f"Train: {X_train['Age'].isnull().sum()}")
print(f"Test: {X_test['Age'].isnull().sum()}")
```

```
def handle_outliers_iqr(train_df, test_df, col):    
    train_df[col] = pd.to_numeric(train_df[col], errors='coerce')
    test_df[col] = pd.to_numeric(test_df[col], errors='coerce')

    # TNH TON THNG S TRN TP TRAIN
    data_train = train_df[col].dropna()
    Q1 = data_train.quantile(0.25)
    Q3 = data_train.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Tnh Median ca tp Train ( thay th)- v data lch
    median_val = data_train.median()
    
    print(f"  - Q1: {Q1} | Q3: {Q3} | IQR: {IQR}")
    print(f"  - Khong an ton (Train): [{lower_bound:.2f}, {upper_bound:.2f}]")
    print(f"  - Gi tr thay th (Median): {median_val}")

    # THAY TH OUTLIER (P DNG CHO C TRAIN V TEST)
    # Nguyn tc: Dng ngng v median ca Train  sa Test
    
    # Hm con  thay th
    def replace_outliers(df, lb, ub, fill_val):
        n_outliers = ((df[col] < lb) | (df[col] > ub)).sum()
        
        # Thay th
        df.loc[(df[col] < lb) | (df[col] > ub), col] = fill_val
        return n_outliers

    count_tr = replace_outliers(train_df, lower_bound, upper_bound, median_val)
    count_te = replace_outliers(test_df, lower_bound, upper_bound, median_val)

    
    return train_df, test_df
X_train, X_test = handle_outliers_iqr(X_train, X_test, 'Work/Study Hours')

print("\nThng k sau khi x l (Train):")
print(X_train['Work/Study Hours'].describe())
```
